```python
import pandas as pd
keywords = pd.read_csv('keywords_new.csv',header=None,names=['A'])
#us_states = us_states.to_dict(orient="list")
key = []
for i in (keywords['A']):
    key.append(i)
key
```

```python
key1 = key[0:29]
key2 = key[29:58]
key3 = key[58:87]
key4 = key[87:]
```

```python
######### All states key words with monthly time ##################

from datetime import date,timedelta,datetime
from dateutil.relativedelta import relativedelta
import pandas as pd                        
from pytrends.request import TrendReq
import os

pytrend = TrendReq(hl='en-US', geo='US', tz=360)
##### X would be custom time range for each month JAN FEB MAR APRIL ###############

x = ['2020-01-01 2020-01-31','2020-02-01 2020-02-29','2020-03-01 2020-03-31',
     '2020-04-01 2020-04-30','2020-05-01 2020-05-28']

###################################################################################

##############keep changing key1,key2,key3,key4 in below for loops for complete data##################

test = 0
for j in x:
    pytrend = TrendReq()    
    pytrend = TrendReq(hl='en-US',geo = 'US', tz=360)
    print("Data Gathering for date : ",j," Done - ",test,"days")
    for i in range(0,len(key4)):
        pytrend.build_payload(
            kw_list=[key4[i]],
            cat=67,
            timeframe= j,
            geo='US',
            gprop='')
        data = pytrend.interest_by_region()
        data = data.reset_index()
        print("pulled for ",i," ",key4[i])
        if i == 0:
            data1 = data
        else:
            data1 = pd.merge(data1, data, on='geoName')
    data1['date'] = str(j)
    if test == 0:
        data2 = data1
    else:
        data2 = data2.append(data1)
    test += 1

print("---------Completed------------")
###############################
# Write file to folder
###############################
#data2.to_csv('C:\\Users\\Anil\\OneDrive\\Desktop\\AnujAdhoc\\region_new_keywords_2020.csv')
```
