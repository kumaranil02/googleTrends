```python
###########################################################################################
# Code for Google searches from  cutomized data range on daily basis by following keywords

# Intrests analyzed by date

#'domestic flight','travel','vacation','resorts','hotels','travel visa','airline ticket',
#'best hotels','airbnb','park'

#For USA region and the category is Travel(67)

###########################################################################################
from datetime import date,timedelta,datetime
from dateutil.relativedelta import relativedelta
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()

#################################
# Start time input
################################
st_year = 2015
st_month = 1
st_day = 1

#################################
# End time input
################################
ed_year = 2020
ed_month = 4
ed_day = 1

x = []
b = date(st_year, st_month, st_day)
while 1 :
    a =  b + relativedelta(months=8)
    net = str(b) + ' ' + str(a)
    if b != date(st_year, st_month, st_day):
        b = b + timedelta(days=1)
        x.append(str(b) + ' ' + str(a))
        b = b - timedelta(days=1)
    else:
        x.append(net)
    if a > date(ed_year, ed_month, ed_day):
        break
    b = b + relativedelta(months=8)

###################################
# pytrends connection request
###################################
pytrend = TrendReq(hl='en-US',geo = 'US', tz=360)

###################################
# Keywords considered for search
# keywords can be changed and added
###################################

keywords = ['domestic flight','travel','vacation','resorts','hotels',
            'travel visa','airline ticket','best hotels','airbnb','park']

key = []
for i in keywords:
    key.append([i])

test = 0
for j in x: 
    print("Data Gathering for date range : ",j," Done - ",test,"months")
    for i in range(0,len(key)):
        pytrend.build_payload(
            kw_list=key[i],
            cat=67,
            timeframe= j,
            geo='US',
            gprop='')
        data = pytrend.interest_over_time()
        data = data.drop(labels=['isPartial'],axis='columns')
        if i == 0:
            data1 = data
        else:
            data1 = pd.merge(data1, data, on='date')
    data1 = data1[sorted([col for col in data1.columns])]
    if test == 0:
        data2 = data1
    else:
        data2 = data2.append(data1)
    test += 9
    
print("---------Completed------------")
###############################
# Write file to folder
###############################
data2.to_csv('C:\\Users\\Anil\\OneDrive\\Desktop\\AnujAdhoc\\daily_5yrs.csv')
```

```python
from datetime import date,timedelta,datetime
from dateutil.relativedelta import relativedelta
import pandas as pd                        
from pytrends.request import TrendReq
import os
#################################
# Start time input
################################
st_year = 2015
st_month = 1
st_day = 1

#################################
# End time input
################################
ed_year = 2020
ed_month = 4
ed_day = 20

pytrend = TrendReq()    
pytrend = TrendReq(hl='en-US',geo = 'US', tz=360)

start_dt = date(st_year, st_month, st_day)
x = []

while start_dt < date(ed_year, ed_month, ed_day) : 
    x.append(str(start_dt))
    start_dt = start_dt + timedelta(days=1)
    

keywords = ['domestic flight','travel','vacation','resorts','hotels',
            'travel visa','airline ticket','best hotels','airbnb','park']

key = []
for i in keywords:
    key.append([i])

test = 0
for j in x:
    pytrend = TrendReq()    
    pytrend = TrendReq(hl='en-US',geo = 'US', tz=360)
    print("Data Gathering for date : ",j," Done - ",test,"days")
    for i in range(0,len(key)):
        pytrend.build_payload(
            kw_list=key[i],
            cat=67,
            timeframe=str(j) + ' ' + str(j),
            geo='US',
            gprop='')
        data = pytrend.interest_by_region()
        data = data.reset_index()
        if i == 0:
            data1 = data
        else:
            data1 = pd.merge(data1, data, on='geoName')
    data1['date'] = str(j)
    if test == 0:
        data2 = data1
    else:
        data2 = data2.append(data1)
    test += 1

print("---------Completed------------")
###############################
# Write file to folder
###############################
data2.to_csv('C:\\Users\\Anil\\OneDrive\\Desktop\\AnujAdhoc\\daily_5yrs_region.csv')
```
