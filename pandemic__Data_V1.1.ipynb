```python
###########################################################################################
# Google searches from data range '2020-01-01 to 2020-04-26' by the following key words

# Intrests analyzed by date

#'domestic flight','travel','vacation','resorts','hotels','travel visa','airline ticket',
#'best hotels','airbnb','park'

#For USA region and the category is Travel(67)

###########################################################################################


import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()

pytrend = TrendReq(hl='en-US',geo = 'US', tz=360)

start_dt = '2020-01-01'

end_dt = '2020-04-26'

keywords = ['domestic flight','travel','vacation','resorts','hotels',
            'travel visa','airline ticket','best hotels','airbnb','park']
key = []
for i in keywords:
    key.append([i])
key

for i in range(0,len(key)):
    pytrend.build_payload(
        kw_list=key[i],
        cat=67,
        timeframe= start_dt + ' ' + end_dt ,
        geo='US',
        gprop='')
    data = pytrend.interest_over_time()
    data = data.drop(labels=['isPartial'],axis='columns')
    if i == 0:
        data1 = data
    else:
        data1 = pd.merge(data1, data, on='date')

data1 = data1[sorted([col for col in data1.columns])]
data1
#data1.to_csv(r'C:\Users\Anil\OneDrive\Desktop\AnujAdhoc\google-topics-by-date1.csv')
```

```python
###########################################################################################
# Google searches from data range '2020-01-01 to 2020-04-26' by the following key words

#Intrests analyzed by region

#'domestic flight','travel','vacation','resorts','hotels','travel visa','airline ticket',
#'best hotels','airbnb','park'

#For USA region and the category is Travel(67)

###########################################################################################

import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()

start_dt = '2020-01-01'

end_dt = '2020-04-26'

pytrend = TrendReq(hl='en-US',geo = 'US', tz=360)

keywords = ['domestic flight','travel','vacation','resorts','hotels',
            'travel visa','airline ticket','best hotels','airbnb','park']
key = []
for i in keywords:
    key.append([i])
    
for i in range(0,len(key)):
    pytrend.build_payload(
        kw_list=key[i],
        cat=67,
        timeframe=start_dt + ' ' + end_dt,
        geo='US',
        gprop='')
    data = pytrend.interest_by_region()
    if i == 0:
        data1 = data
    else:
        data1 = pd.merge(data1, data, on='geoName')

data1  = data1[sorted([col for col in data1.columns])]
data1['date'] = end_dt
#data1.to_csv(r'C:\Users\Anil\OneDrive\Desktop\AnujAdhoc\google-topics-by-region1.csv')
```
