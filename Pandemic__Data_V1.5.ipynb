```python
###########################################################################################
# Google searches from data range '2020-01-01 to 2020-04-26' by the following key words
`
# Intrests analyzed by date

#'domestic flight','travel','vacation','resorts','hotels','travel visa','airline ticket',
#'best hotels','airbnb','park'

#For USA region and the category is Travel(67)

###########################################################################################
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()

pytrend = TrendReq(hl='en-US',geo = 'US', tz=360)

start_dt = '2020-01-01'

end_dt = '2020-04-29'

keywords = ['domestic flight','travel','vacation','resorts','hotels',
            'travel visa','airline ticket','best hotels','airbnb','park']
key = []
for i in keywords:
    key.append([i])
key

for i in range(0,len(key)):
    pytrend.build_payload(
        kw_list=key[i],
        cat=67,
        timeframe= start_dt + ' ' + end_dt ,
        geo='US',
        gprop='')
    data = pytrend.interest_over_time()
    data = data.drop(labels=['isPartial'],axis='columns')
    if i == 0:
        data1 = data
    else:
        data1 = pd.merge(data1, data, on='date')

data1 = data1[sorted([col for col in data1.columns])]


############################################################
#collecting last update date to add in region wise report
#
############################################################
data1 = data1.reset_index()
date_to_add = str(data1['date'][len(data1['date']) - 1]).split()
date_to_add = date_to_add[0]


############################################################
#Data written to todays folder
############################################################
file_name = '\google-topics-by-date_' + str(today) + '.csv'
data1.to_csv(path+file_name)
```

```python
#################################################################################
#
# Automatic zip file email delivery
#
#################################################################################
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import os.path

os.chdir(r'C:\Users\Anil\OneDrive\Desktop\AnujAdhoc\Covid_Daily_Data')
os.makedirs(name = str(today))
path = os.getcwd()+ "\\" + str(today)
os.chdir(path)
path = os.getcwd()


def send_email(email_recipient,
               email_subject,
               email_message,
               attachment_location = ''):

    email_sender = '- Senders - @gmail.com'

    msg = MIMEMultipart()
    msg['From'] = email_sender
    msg['To'] = email_recipient
    msg['Subject'] = email_subject

    msg.attach(MIMEText(email_message, 'plain'))

    if attachment_location != '':
        filename = os.path.basename(attachment_location)
        attachment = open(attachment_location, "rb")
        part = MIMEBase('application', 'octet-stream')
        part.set_payload(attachment.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition',
                        "attachment; filename= %s" % filename)
        msg.attach(part)

    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.ehlo()
        server.starttls()
        server.login('my-login-email-id@xxx.com', "passwordXXXX")
        text = msg.as_string()
        server.sendmail(email_sender, email_recipient, text)
        print('email sent')
        server.quit()
    except:
        print("SMPT server connection error")
    return True

send_email('email1@xxx.com,email2@xxx.com,email3@xxx.com',
           'Daily Covid Data',
           'Team,Please consider the attached file for recent data', 
           path)
```
